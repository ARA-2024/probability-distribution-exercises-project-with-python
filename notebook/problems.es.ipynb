# Problemas de distribución de probabilidad

## Distribución binomial

En teoría de la probabilidad y estadística, la distribución binomial es la distribución de probabilidad discreta que da sólo dos resultados posibles en un experimento; por ejemplo, al lanzar una moneda, puede salir cara o cruz.

```python
from scipy.stats import binom

# Definir todas las funciones de probabilidad relacionadas con esta distribución.

def dbinom(x, size, prob = 0.5):
    """
    Calcula la estimación puntual de la distribución binomial.
    """
    result = binom.pmf(k = x, n = size, p = prob, loc = 0)

    return result

def pbinom(q, size, prob = 0.5):
    """
    Calcula el acumulado de la distribución binomial.
    """
    result = binom.cdf(k = q, n = size, p = prob, loc = 0)

    return result

def qbinom(p, size, prob = 0.5):
    """
    Calcula la función cuantil a partir de la distribución binomial.
    """
    result = binom.ppf(q = p, n = size, p = prob, loc = 0)

    return result

def rbinom(n, size, prob = 0.5):
    """
    Genera variables aleatorias a partir de la distribución binomial.
    """
    result = binom.rvs(n = size, p = prob, size = n)

    return result
```

```python
# Distribución binomial(10, 0.2), dónde n = 10 es el número de ensayos y p = 0.2 es la probabilidad de éxito en un solo ensayo
import numpy as np

np.random.seed(42)

print(f"Probability that a Binomial(10, 0.2) takes the value 2: {dbinom(2, size = 10, prob = 0.2)}")
print(f"Probability that a Binomial(10, 0.2) will take a value less than 2: {pbinom(2, size = 10, prob = 0.2)}")
print(f"Which value of a Binomial(10, 0.2) has a cumulative probability of 0.9? {qbinom(0.9, size = 10, prob = 0.2)}")
print(f"Generate 2000 random values from a Binomial(10, 0.2) distribution: {rbinom(2000, size = 10, prob = 0.2)}")
```

```python
import matplotlib.pyplot as plt

np.random.seed(42)

size = 10
prob = 0.2
data = rbinom(2000, size = 10, prob = 0.2)

pmf = dbinom(data, size = size, prob = prob)
cdf = pbinom(data, size = size, prob = prob)
ppf = qbinom(data, size = size, prob = prob)

fig, axis = plt.subplots(3, 1, figsize = (5, 7))

axis[0].stem(data, pmf, "r-")
axis[1].stem(data, cdf, "b-")
axis[2].stem(data, ppf, "g-")

# Adjust the layout
plt.tight_layout()

# Show the plot
plt.show()
```

Escriba las funciones anteriores pero para distribuciones normales y uniformes.

```python
import numpy as np
import scipy
from scipy import stats
from scipy.stats import norm
import matplotlib.pyplot as plt

# Distribución normal de media 3 y desviación estándar 2
# Se definen todas las funciones de probabilidad relacionadas con esta distribución.
## Probability density function
def normal_pdf(x, loc, scale):
     return stats.norm.pdf(x, loc, scale)
## Cumulative distribution function
def normal_cdf(x, loc, scale):
     return stats.norm.cdf(x, loc, scale)
## Percentil of the normal distribution
def normal_q(q, loc, scale):
     return stats.norm.ppf(q, loc, scale)
## Variables aleatorias de distribución normal
def normal_values(loc, scale, size):
     return stats.norm.rvs(loc, scale, size)

print(f"PDF of a Normal Distribution with mean=3 and sigma=2 for the value 4 is: {normal_pdf(4, loc=3, scale=2)}")
print(f"CDF of a Normal Distribution with mean=3 and sigma=2 for the value 4 is: {normal_cdf(4, loc=3, scale=2)}")
print(f"Which value of a Normal Distribution with mean=3 and sigma=2 is the 95% percentile? {normal_q (0.95, loc=3, scale=2)}")
print(f"Generate 100 random values from a Normal Distribution with mean=3 and sigma=2: {normal_values(loc=3, scale=2, size=(10, 10))}")

# Representamos las funciones anteriores
data = np.linspace(-10, 10, 100)
pdf = normal_pdf(data, loc=3, scale=2)
cdf = normal_cdf(data, loc=3, scale=2)
ppf = normal_q(data, loc=3, scale=2)

fig, axis = plt.subplots(3, 1, figsize = (5, 7))

axis[0].stem(data, pdf, "r-")
axis[1].stem(data, cdf, "b-")
axis[2].stem(data, ppf, "g-")

# Adjust the layout
plt.tight_layout()

# Show the plot
plt.show()


# Distribución uniforme de valor mínimo 1 y scale 9
# Se definen todas las funciones de probabilidad relacionadas con esta distribución.
## Probability density function
def uniform_pdf(x, loc, scale):
     return stats.uniform.pdf(x, loc, scale)
## Cumulative distribution function
def uniform_cdf(x, loc, scale):
     return stats.uniform.cdf(x, loc, scale)
## Percentil of the uniform distribution
def uniform_q(q, loc, scale):
     return stats.uniform.ppf(q, loc, scale)
## Variables aleatorias de distribución uniforme
def uniform_values(loc, scale, size):
     return stats.uniform.rvs(loc, scale, size)

print(f"PDF of an Uniform Distribution with loc=1 and scale=9 for the value 4 is: {uniform_pdf(4, loc=1, scale=9)}")
print(f"CDF of an Uniform Distribution with loc=1 and scale=9 for the value 4 is: {uniform_cdf(4, loc=1, scale=9)}")
print(f"Which value of an Uniform Distribution with loc=1 and scale=9 is the 95% percentile? {uniform_q (0.95, loc=1, scale=9)}")
print(f"Generate 100 random values from an Uniform Distribution with loc=1 and scale=9: {uniform_values(loc=1, scale=9, size=10)}")

# Representamos las funciones anteriores
data = np.linspace(0, 10, 100)
pdf = uniform_pdf(data, loc=1, scale=9)
cdf = uniform_cdf(data, loc=1, scale=9)
ppf = uniform_q(data, loc=1, scale=9)

fig, axis = plt.subplots(3, 1, figsize = (5, 7))

axis[0].stem(data, pdf, "r-")
axis[1].stem(data, cdf, "b-")
axis[2].stem(data, ppf, "g-")

# Adjust the layout
plt.tight_layout()

# Show the plot
plt.show()
```
